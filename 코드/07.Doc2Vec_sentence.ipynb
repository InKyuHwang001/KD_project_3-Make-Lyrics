```python
df123 = pd.read_pickle('/content/drive/MyDrive/5조_파이널PJT/코드/data/doc2vec.pkl')
df123
```

```python
df123 = df123.reset_index()
df123['title_num'] = df123['title_artist'] +'_' + df123['index'].astype('str')
df123
```

```python
from gensim.models.doc2vec import TaggedDocument
from tqdm import tqdm
```

```python
tagged_corpus_list = []

for index, row in tqdm(df123.iterrows(), total=len(df123)):
  text = row['lyric']
  tag = row['title_num']
  tagged_corpus_list.append(TaggedDocument(tags=[tag], words=mecab.morphs(text)))

print('문서의 수 :', len(tagged_corpus_list))
```

```python
AI_text = "너와 함께 하고 싶은 일들에서"     # 너와 함께 하고 싶은 일들을 , 조사 치환
AI_text2 = "그 사람 향기가 자꾸 생각이 난다" # 그 사람 손길이 자꾸 생각이 난다, 단어 치환
AI_text3 = "하지마 음주운전"    # 하지마 하지마 음주운전 하지마
AI_text4 = "소중한 걸 난 찾고 있어"   # 소중한 무언가를 난 또 쫓고 있어
AI_text5 = "별빛이 내리는 끝없던 밤"   # 끝없이 별빛이 내리던 밤
```

```python
tagged_corpus_list.append(TaggedDocument(tags=["AI가사"], words=mecab.morphs(AI_text)))
tagged_corpus_list.append(TaggedDocument(tags=["AI가사2"], words=mecab.morphs(AI_text2)))
tagged_corpus_list.append(TaggedDocument(tags=["AI가사3"], words=mecab.morphs(AI_text3)))
tagged_corpus_list.append(TaggedDocument(tags=["AI가사4"], words=mecab.morphs(AI_text4)))
tagged_corpus_list.append(TaggedDocument(tags=["AI가사5"], words=mecab.morphs(AI_text5)))
```

```python
tagged_corpus_list[-5:]
```

# 2. Doc2Vec 학습 및 테스트

```python
from gensim.models import doc2vec
```

```python
model = doc2vec.Doc2Vec(vector_size=20, alpha=0.025, min_alpha=0.025, workers=8, window=8)

# Vocabulary 빌드
model.build_vocab(tagged_corpus_list)
print(f"Tag Size: {len(model.docvecs.doctags.keys())}", end=' / ')

# Doc2Vec 학습
model.train(tagged_corpus_list, total_examples=model.corpus_count, epochs=20)

# 모델 저장
model.save('/content/drive/MyDrive/5조_파이널PJT/코드/model_result/lyric123.sent2vec')
```

```python
similar_doc = model.docvecs.most_similar('AI가사')
print(similar_doc)
```
